# Beam Katas
## ParDo OneToMany

Kata: Please write a ParDo that maps each input sentence into words tokenized by whitespace (" ").


 Hint 1 
Override process method. You can return an Iterable for multiple elements or call "yield" for each element to return a generator.

 Hint 2 
Use ParDo with DoFn.

 Hint 3 
Refer to the Beam Programming Guide "ParDo" section for more information.

```
!pip install apache-beam -qqq

import apache_beam as beam
from apache_beam.runners.interactive import interactive_runner
```

## Python Collection

```
class BreakIntoWordsDoFn(beam.DoFn):

    def process(self, element):
        return element.split()
      
['Hello Beam', 'It is awesome'] | beam.ParDo(BreakIntoWordsDoFn())
```

## Beam Pcollection

```
p = beam.Pipeline(interactive_runner.InteractiveRunner())

class BreakIntoWordsDoFn(beam.DoFn):

    def process(self, element):
        return element.split()
            
(p | 'Create' >> beam.Create(['Hello Beam', 'It is awesome'])
   | 'BreakIntoWords' >> beam.ParDo(BreakIntoWordsDoFn()))

p.run()
```
